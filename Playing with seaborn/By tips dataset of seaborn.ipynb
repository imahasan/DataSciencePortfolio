```python
import seaborn as sns

# We also need matplotlib to visualize the data
import matplotlib.pyplot as plt

# Below command for visualization of the plots instantly
%matplotlib inline
```

Lets bring the "tips" dataset and also call it tips
Then we will check the header of the dataset to have an idea of the data it contains

```python
tips = sns.load_dataset("tips")
tips.head()
```

Lets create a barplot first where we compare the tip amount in the days of the week to know which day generates most amount of tips!!!

```python
# Lets choose a color palette for out plot also
sns.barplot(data=tips, x='day', y='tip', estimator='sum' ,errorbar=None, palette='coolwarm')
# We wil also set a title of our plot
plt.title('Daily Tips in ($)')
```

So we generate most tips on Saturday and Sundays! Excelent!

Lets see who gives us most tips! Boys or Girls!

```python
sns.barplot(data=tips, x='day', y='tip', estimator='sum' ,errorbar=None, palette='coolwarm',hue='sex')

plt.title('Daily Tips in ($)')
```

So, we generate most tips from the Males!

Lets see when we generate the most tips!

```python
sns.barplot(data=tips, x='time', y='tip', estimator='sum' ,errorbar=None, palette='coolwarm')

plt.title('Daily Tips in ($)')
```

It's dinner time!
